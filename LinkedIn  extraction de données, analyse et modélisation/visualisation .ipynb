```python

Edu_df=df[["profile","ed_date","ed_deg","ed_title"]]
Edu_df
```

```python

# Function to identify degree
def degree(x):
    if x.lower().find('master') != -1 or x.lower().find('ms') != -1 or x.lower().find('m.s') != -1 or x.lower().find("m1") != -1 or x.lower().find('m2') != -1 or x.lower().find("master's") != -1:
        return(1)
    if x.lower().find('phd') != -1 or x.lower().find('doctorales') != -1 or x.lower().find('thèse') != -1 or x.lower().find('researcher') != -1 or x.lower().find('phd') != -1 or x.lower().find('ph.d') != -1 or x.lower().find('docteur') != -1:
        return(2)
    if x.lower().find('bachelor') != -1 or x.lower().find('licence ') != -1 or x.lower().find('bs') != -1 or x.lower().find('bac +3') != -1 or x.lower().find('b.s') != -1:
        return(3)
    if x.lower().find('engineering') != -1 or x.lower().find('engineer') != -1 or x.lower().find("engineer's") != -1 or x.lower().find('ingénieurie') != -1 or x.lower().find('ingénieure') != -1 or x.lower().find('ingénieur') != -1 or x.lower().find("d’ingénieur") != -1:
        return(4)                                                                                                                 
    else:
        return(0)
```

```python
Edu_df['deg'] = list(map(degree, Edu_df['ed_deg']))
```

```python
def title_master(x):
    if x.lower().find('data science') != -1 or x.lower().find('sciences des données	') != -1 :
        return(1)
    if x.lower().find('stat') != -1 or x.lower().find('math') != -1 or x.lower().find('phy') != -1:
        return(2)
    if x.lower().find('management') != -1 or x.lower().find('admin') != -1 or x.lower().find('supply') != -1:
        return(3)
    if x.lower().find('business') != -1 or x.lower().find('analytics ') != -1 or x.lower().find('big data') != -1 or x.lower().find('cloud') != -1 or x.lower().find('internet of things') != -1 :
        return(4)
    if x.lower().find('software') != -1 or x.lower().find('engineering') != -1:
        return(5)
    if x.lower().find('elect') != -1:
        return(6)
    else:
        return(0)

    
# Create degree column
edum['deg_title'] = list(map(title_master, edum['ed_title']))

#Took out values = 0 because they are only gpa scores 
edum = edum[edum.deg_title != 0]
```

```python
def experience(x):
    if x.lower().find('employee') != -1 or x.lower().find('cdi') != -1 or x.lower().find('cdd') != -1 or x.lower().find('temps plein') != -1 or x.lower().find('full-time') != -1 or x.lower().find('temps partiel') != -1 or x.lower().find('part-time') != -1:
        return(1)
    if x.lower().find('stage') != -1 or x.lower().find('internship') != -1:
        return(2)
    if x.lower().find('indépendant') != -1 or x.lower().find('freelance') != -1 or x.lower().find('self-employed') != -1:
        return(3)
    else:
        return(0)
```

```python
exp['deg_experience'] = list(map(experience, exp['type_experience']))
```

```python
exp
```

```python
experiencelbar = exp['deg_experience'].value_counts()
experiencelbar
```

```python
# library
import matplotlib.pyplot as plt
 
# create data
names = ['employee', 'stage', 'freelance', 'other']
size = experiencelbar
 
# Create a circle at the center of the plot
my_circle = plt.Circle( (0,0), 0.7, color='white')

# Give color names
plt.pie(size, labels=names, colors=['red','green','blue','skyblue'])
p = plt.gcf()
p.gca().add_artist(my_circle)

# Show the graph
plt.show()
```
